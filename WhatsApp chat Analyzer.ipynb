```python
import re
import pandas as pd
import numpy as np
```

```python
f = open(r"D:\University_Project\WhatsApp Chat with Data Sc+Computer Sc.txt",'r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
```

```python
dates = re.findall(pattern, data)
```

```python
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df
```

```python
pd.to_datetime(df['message_date'], format="%d/%m/%y, %H:%M - ")
```

```python
df['date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M - ')
# Use dt.strftime() to format datetime objects
df['formatted_date'] = df['date'].dt.strftime('%Y-%m-%d %H:%M:%S')

df
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
# df.drop(columns=['user_message'], inplace=True)
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
period = []
for hour in df['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str(hour + 1))
    else:
        period.append(str(hour) + "-" + str(hour + 1))
df['period'] = period
return df
```
